#loading the data - image

```python
img = cv2.imread('./red_white_balls.png') # the red color object in this image turns to blue(mostly) after reading it. Tt's color is changing
show(img)
```

```python
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
show(img)
```

Normalizing the image

```python
max_dim = max(img.shape) # =700
print(max_dim)
scale = 700/max_dim
img = cv2.resize(img,None, fx=scale, fy=scale)
```

GaussianBlurring <br>  Link:https://docs.opencv.org/3.1.0/d4/d13/tutorial_py_filtering.html<br>
In this, instead of box filter, gaussian kernel is used. It is done with the function, cv2.GaussianBlur(). We should specify the width and height of kernel which should be positive and odd. We also should specify the standard deviation in X and Y direction, sigmaX and sigmaY respectively. If only sigmaX is specified, sigmaY is taken as same as sigmaX. If both are given as zeros, they are calculated from kernel size. Gaussian blurring is highly effective in removing gaussian noise from the image.

If you want, you can create a Gaussian kernel with the function, cv2.getGaussianKernel().

The above code can be modified for Gaussian blurring:

    blur = cv2.GaussianBlur(img,(5,5),0)

```python
#BLurring the image to reduce noise and hence normalize it better 
img_blur = cv2.GaussianBlur(img, (7,7),0)
#Converting the image from rgb to hsv to extract more charateristics of the image e.g. color, intensity, shades etc.
img_blur_hsv = cv2.cvtColor(img_blur, cv2.COLOR_RGB2HSV)
```

```python
show(img_blur_hsv)
```

Preparing a mask below:<br>
define lower and upper color-bound of the color object you want to detect <br>see the documentation link - https://docs.opencv.org/3.0-beta/doc/py_tutorials/py_imgproc/py_colorspaces/py_colorspaces.html 
<br>
in our case we are simply detecting a red color object on the basis of color-spectrum and the brightness

```python
#mask on the basis of color factor
min_red = np.array([0,100,80])
max_red = np.array([10,256,256])
#Mask1  = cv2.inRange(img_blur_hsv, min_red, max_red)
Mask1  = cv2.inRange(img_blur_hsv, min_red, max_red)

#mask on the basis of brightnss factor
min_redb = np.array([170,80,80])
max_redb = np.array([180,256,256])
#Mask2 = cv2.inRange(img_blur_hsv, min_redb, max_redb)
Mask2 = cv2.inRange(img_blur_hsv, min_redb, max_redb)
#concatenate both masks for feature extraction 
mask = Mask1+ Mask2
```

```python

# mycv2imshow('mask1',Mask1)
# mycv2imshow('mask2',Mask2)
# mycv2imshow('mask',mask)
# # cv2.waitKey()
# cv2.destroyAllWindows()
show(Mask1) #'Mask1',
show(Mask2) #'Mask2'
show(mask)  #'mask',
```

post-processing of mask i.e. cleaning process (mask_clean) of mask as it may contain noises<br> LINK to DOC. : https://docs.opencv.org/3.0-beta/doc/py_tutorials/py_imgproc/py_morphological_ops/py_morphological_ops.html <br>
creating kernels->applying morphological operations below:


```python
#creating an ellipse shaped kernels
kernel  = cv2.getStructuringElement(cv2.MORPH_ELLIPSE, (5,5))
#applying closing operation :  Dilation followed by Erosion   - erosion is used to whiten the foreground and dilation works opposite to erosion.
mask_closing = cv2.morphologyEx(mask, cv2.MORPH_CLOSE, kernel)
#applying opening operation - opposite of closing operation  - used to remove noise, using Dilation, from the foreground caused by Erosion 
mask_clean = cv2.morphologyEx(mask, cv2.MORPH_OPEN, kernel)

# cv2.startWindowThread()
# cv2.imshow('mask closing', mask_closing)
# cv2.waitKey()
# cv2.destroyAllWindows()
show(mask_closing)
show(mask_clean)
```

```python
# Extract biggest bounding box
#big_contour, red_mask = find_biggest_contour(mask_clean)
# #Extract biggest bounding box
biggest_contour, mask_red = find_biggest_contour(mask_clean)
```

```python
#Apply 
overlay = overlay_mask(mask_red, img )
```

```python
circled = circle_contour_ellipse(overlay, biggest_contour)

show(circled)
```
